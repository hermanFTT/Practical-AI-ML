### Read in an audio file using Librosa

First we'll load audio using one approach, that is, to use the Librosa library. Librosa has many functions to do audio processing tasks, one of them is the ```.load()``` function.

[Check out the API](https://librosa.org/doc/latest/generated/librosa.load.html)

Here we specify the file name, and that ```sr==None```, why?

````python
audio_amps, audio_sample_rate = librosa.load('water.wav', sr=None)
````

Let's take a look what is inside ```audio_amps```

W1_Monday_Slicing.jpg


2.   Write some code that will extract the first 0.5 seconds from ```audio_short```.
3.   To do this, you will slicing the Python array, for example: ```array[0:100]```.
4.   Use the sampling rate to help you. Remember, in our example, there are 22050 data points in 1 second of time! The sampling rate was stored in the variable ```sr```. You can't have decimal values as indices when slicing a Python array, so you might need to type cast: ```int()```, for example: ```array[0:int(50/2)]```.
5.   Store the result of your clip in a variable called ```middle_first_part```.

<br><br><br><br><br><br><br><br><br><br><br><br><br>

````python
# Write some code that will extract the first 0.5 seconds from audio_short.
middle_first_part = audio_short[0:int(sr/2)]
````

````python
# Check the length of your clip, it should be 0.5.
len(middle_first_part)/sr
````

````python
# Play the audio clip.
Audio(middle_first_part, rate=sr)
````

3.   Store the result in a variable called ```full_audio```.
4.   Print the length of the new audio, it should be 4.87

6.   Check that the new audio is the same as the original one. You can use Numpy's ```array_equiv()``` function. [Check the API.](https://numpy.org/doc/stable/reference/generated/numpy.array_equiv.html) The result from this should be ```True```.
